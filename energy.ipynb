```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm

energy=pd.read_csv(r'C:\Users\Lisa\Downloads\energydata.csv')
```

```python
energy.head(10)
```

```python
#remove World from Country column
energy2=energy.loc[energy['Country']!='World']
```

```python
#pivot energy types rows for energy production column to columns
energy_production=pd.pivot_table(energy2,values='Energy_production',index=['Country','Year'],columns='Energy_type')
```

```python
#rename pivoted columns
energy_production_renamed=energy_production.rename(columns={'all_energy_types':'Production_all_energy_types','coal':'Production_coal','natural_gas':'Production_nat_gas','nuclear':'Production_nuclear','petroleum_n_other_liquids':'Production_petroleum','renewables_n_other':'Production_renewables'})
```

```python
#pivot energy types rows from energy consumption column to columns
energy_consumption=pd.pivot_table(energy2,values='Energy_consumption',index=['Country','Year'],columns='Energy_type')
```

```python
#rename pivoted columns
energy_consumption_renamed=energy_consumption.rename(columns={'all_energy_types':'Consumption_all_energy_types','coal':'Consumption_coal','natural_gas':'Consumption_nat_gas','nuclear':'Consumption_nuclear','petroleum_n_other_liquids':'Consumption_petroleum','renewables_n_other':'Consumption_renewables'})
```

```python
#subset columns from original df
energy_othercols=energy2[['Country','Year','GDP','Population','Energy_intensity_per_capita','Energy_intensity_by_GDP','CO2_emission']]
```

```python
#join pivoted energy production df with other columns from original df
merged_energy=pd.merge(energy_production_renamed, energy_othercols, how='left', left_on=['Country','Year'],right_on=['Country','Year'])
```

```python
#join pivoted energy consumption df with merged df
merged_energy_2=pd.merge(merged_energy,energy_consumption_renamed, on=['Country','Year'],how='inner')
cleanedup_energy=merged_energy_2.drop_duplicates()
cleanedup_energy.shape
cleanedup_energy.head(20)
```

```python
print(list(merged_energy_2.columns))
```

```python
#group by country and year and aggregate by mean of all other columns
energy_grouped = merged_energy_2.groupby(['Country','Year']).aggregate({'Production_all_energy_types':'mean','Production_coal':'mean','Production_nat_gas':'mean', 'Production_nuclear':'mean', 'Production_petroleum':'mean', 'Production_renewables':'mean', 'GDP':'mean', 'Population':'mean', 'Energy_intensity_per_capita':'mean', 'Energy_intensity_by_GDP':'mean', 'CO2_emission':'mean', 'Consumption_all_energy_types':'mean', 'Consumption_coal':'mean', 'Consumption_nat_gas':'mean', 'Consumption_nuclear':'mean', 'Consumption_petroleum':'mean', 'Consumption_renewables':'mean'}).reset_index()
energy_grouped.shape
```

```python
energy_counted = cleanedup_energy.groupby(['Country','Year'])['Production_all_energy_types'].count()
energy_counted_df=energy_counted.to_frame()
energy_counted_df_sorted=energy_counted_df.sort_values('Production_all_energy_types',ascending=False)
energy_counted_df['Production_all_energy_types'].value_counts()
```

```python
#correlations among all columns/variables
correlation=energy_grouped.corr()
print(correlation)
```

```python
#correlations with renewable energy production
correlation_prod_renewables=correlation[['Production_renewables']].sort_values(by='Production_renewables',ascending=False)
print(correlation_prod_renewables)
```

```python
final_energy=energy_grouped.dropna(subset=['Consumption_all_energy_types','GDP','CO2_emission'])
final_energy.head(20)
#Is it ok that there are multiple and different values for the same country/Year combination?
```

```python
independent_variables=final_energy[['Consumption_all_energy_types','GDP','CO2_emission']]
dependent_variable=final_energy['Production_renewables']
dependent_variable=dependent_variable.to_frame()
independent_variables=sm.add_constant(independent_variables)
regression_model_science=sm.OLS(dependent_variable,independent_variables).fit()
print(regression_model_science.summary())
```
